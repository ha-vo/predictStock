```python
from yahoo_fin.stock_info import get_data
amazon_weekly= get_data("amzn", start_date="12/04/2009", end_date="15/03/2024", index_as_date = True, interval="1d")
amazon_weekly
```

```python
series_shift = amazon_weekly['close'].shift()
amazon_weekly['pre_close'] = series_shift
amazon_weekly['pre_open'] = amazon_weekly['open'].shift()
```

```python
amazon_weekly
```

```python
amazon_weekly.dropna(inplace=True)
amazon_weekly
```

# predict open price

```python
X = amazon_weekly[["pre_close","pre_open"]].values
y = amazon_weekly["open"].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 32, test_size = 1/4)
```

```python
ln = LinearRegression().fit(X_train, y_train)
```

```python
y_pred = ln.predict(X_test)
```

```python
for i in range(20):
    print(y_test[i], y_pred[i])
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
print('Coefficient of determination (R^2): %.2f' % r2_score(y_test, y_pred))
```

## predict the close price

```python
X = amazon_weekly[["pre_close","open"]].values
y = amazon_weekly["close"].values
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 32, test_size = 1/4)
ln = LinearRegression().fit(X_train, y_train)
y_pred = ln.predict(X_test)

for i in range(20):
    print(y_test[i], y_pred[i])
```

```python
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
print('Coefficient of determination (R^2): %.2f' % r2_score(y_test, y_pred))
```
